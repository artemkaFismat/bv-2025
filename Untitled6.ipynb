```python
from llama_cpp import Llama

llm = Llama.from_pretrained(
  repo_id="QuantFactory/T-lite-0.1-GGUF",
	filename="T-lite-0.1.Q8_0.gguf",
    n_ctx=8192,             # Context length to use
    #n_threads=10,          # Number of CPU threads to use
    n_gpu_layers=-1,        # Number of model layers to offload to GPU
    verbose=True
)
```

```python
promt = "Язык программирования Тривиль"
output = llm.create_chat_completion(
	messages = [
     {"role": "system",
      "content": "Ты известный ученый с мировым именем. Отвечай как доктор наук. Ответ верни в формате JSON"},
     {"role": "user",
      "content": """## Анализ
                    Проанализируйте тему: Конструирование компиляторов для цифровых вычислительных машин
                    ## Задача
                    Пожалуйста, составьте план с соответствующими заголовками разделов и укажите 2-3 ключевых момента в каждом разделе. План должен соответствовать стандартам академического письма и подходить для публикации в рецензируемом журнале. Четко действуй в соответствии с инструкцией.
                    ## Инструкция
                    Убедитесь, что:
                    - Используйте понятный академический язык
                    - Включите соответствующие цитаты
                    - Следуйте стандартной структуре IMRaD:
                            1. Введение — описание характера и масштабов проблемы, а также её изученности.
                            2. Методы — описание хода исследования и применённых методов.
                            3. Результаты — основные результаты изысканий.
                            4. Обсуждение — анализ и оценка результатов в контексте существующей научной литературы.
                            5. Выводы — резюме по анализу результатов проведённого исследования и перспективы.
                    - Решите проблемы, которые могут возникнуть у рецензента
                    - Подчеркните новизну и значимость исследования
                    ## Формат вывода результата
                    Результатом должен стать подробный план без списка литературы, который может послужить основой для написания полноценной научной статьи. План статьи представь в формате JSON {"title":"<заголовок раздела>", "description":"<ключевые моменты раздела>"}
                    ## Проверка результата
                    Проверь, что:
                    - Разделы соответствуют заданной структуре
                    - Заголовки разделов не повторяются
                    - Заголовки разделов на русском языке
                    - Ключевые моменты разделов содержательны
                    - Результат вывода соответствует формату JSON
                """
      }

  ]
)
print(output)
```
